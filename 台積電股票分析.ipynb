載入台積電原始資料

```python
import pandas as pd

df = pd.read_csv('2330.csv', encoding = 'big5')
df.tail(20)
```

每個欄位名稱逐行列出

```python
for col in df.columns:
    print(col)
```

抓取特定欄位，輸出到df1

```python
# 指定需要的欄位
columns_to_select = [
    '年月日',
    '開盤價(元)',
    '最高價(元)',
    '最低價(元)',
    '收盤價(元)',
    '成交量(千股)',
    '成交值(千元)',
    '報酬率％',
    '5日均價(元)',
    '10日均價(元)',
    '15日均價(元)',
    '20日均價(元)',
    '當日均價(元)',
    '3日均價(元)'
]

# 從df中選取指定的欄位
df1 = df[columns_to_select]

# 顯示選取後的資料
df1
```

新增當日是否漲跌欄位

```python
# 創建 '當日漲跌' 欄位，並儲存到新的 DataFrame df2
df2 = df1.copy()
df2.loc[:, '當日漲跌'] = df2.apply(
    lambda row: 1 if row['收盤價(元)'] > row['開盤價(元)']
                else (2 if row['收盤價(元)'] < row['開盤價(元)'] else 0),
    axis=1
)
df2
```

檢查是否有空值

```python
# 檢查每個欄位的空值數量
null_values = df2.isnull().sum()

# 顯示結果
print(null_values)

# 只查看有空值的欄位
print(null_values[null_values > 0])
```

轉換日期格式：將日期轉換為 datetime 格式，便於進行時間序列分析。

```python
df2['年月日'] = pd.to_datetime(df2['年月日'], format='%Y/%m/%d')
```

開盤價(元), 最高價(元), 最低價(元), 收盤價(元) \\

```python
df2[['開盤價(元)', '最高價(元)', '最低價(元)', '收盤價(元)']] = df2[['開盤價(元)', '最高價(元)', '最低價(元)', '收盤價(元)']].apply(pd.to_numeric, errors='coerce')
```

成交量(千股), 成交值(千元)

```python
df2['成交量(千股)'] = df2['成交量(千股)'].str.replace(',', '').astype(float)
df2['成交值(千元)'] = df2['成交值(千元)'].str.replace(',', '').astype(float)
```

### 報酬率％
移除百分比符號並轉換格式：與 漲跌幅 類似，將百分比轉換為浮點數格式。

```python
df2['報酬率％'] = df2['報酬率％'].astype(float)
```

5日均價(元), 10日均價(元), 15日均價(元), 20日均價(元), 當日均價(元), 3日均價(元)


```python
df2[['5日均價(元)', '10日均價(元)', '15日均價(元)', '20日均價(元)', '當日均價(元)', '3日均價(元)']] = df2[['5日均價(元)', '10日均價(元)', '15日均價(元)', '20日均價(元)', '當日均價(元)', '3日均價(元)']].apply(pd.to_numeric, errors='coerce')
```

對數值型欄位進行 標準化 或 正規化，使模型能更好地收斂

針對df2與df3進行機器學習

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 處理 df2
#X_df2 = df2[['開盤價(元)', '最高價(元)', '最低價(元)', '收盤價(元)', '成交量(千股)', '成交值(千元)', '報酬率％',
#             '5日均價(元)', '10日均價(元)', '15日均價(元)', '20日均價(元)', '當日均價(元)', '3日均價(元)']]

X_df2 = df2[['開盤價(元)', '最高價(元)', '最低價(元)', '收盤價(元)', '報酬率％',
             '5日均價(元)', '10日均價(元)', '15日均價(元)', '當日均價(元)', '3日均價(元)']]

y_df2 = df2['當日漲跌']

# 處理 df3
X_df3 = df3[['開盤價(元)', '最高價(元)', '最低價(元)', '收盤價(元)', '報酬率％',
             '5日均價(元)', '10日均價(元)', '15日均價(元)', '當日均價(元)', '3日均價(元)', '情感欄位']]
y_df3 = df3['當日漲跌']

# 分割數據集為訓練集和測試集
X_train_df2, X_test_df2, y_train_df2, y_test_df2 = train_test_split(X_df2, y_df2, test_size=0.1, random_state=42)
X_train_df3, X_test_df3, y_train_df3, y_test_df3 = train_test_split(X_df3, y_df3, test_size=0.1, random_state=42)

# 創建決策樹模型
clf_df2 = DecisionTreeClassifier(random_state=42)
clf_df3 = DecisionTreeClassifier(random_state=42)

# 訓練模型
clf_df2.fit(X_train_df2, y_train_df2)
clf_df3.fit(X_train_df3, y_train_df3)

# 預測測試集
y_pred_df2 = clf_df2.predict(X_test_df2)
y_pred_df3 = clf_df3.predict(X_test_df3)

# 評估模型
accuracy_df2 = accuracy_score(y_test_df2, y_pred_df2)
accuracy_df3 = accuracy_score(y_test_df3, y_pred_df3)

print("df2 決策樹模型準確度:", accuracy_df2)
print("df3 決策樹模型準確度:", accuracy_df3)
```

隨機森林

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 處理 df2
X_df2 = df2[['開盤價(元)', '最高價(元)', '最低價(元)', '收盤價(元)', '報酬率％',
             '5日均價(元)', '10日均價(元)', '15日均價(元)', '當日均價(元)', '3日均價(元)']]
y_df2 = df2['當日漲跌']

# 處理 df3
X_df3 = df3[['開盤價(元)', '最高價(元)', '最低價(元)', '收盤價(元)', '報酬率％',
             '5日均價(元)', '10日均價(元)', '15日均價(元)', '當日均價(元)', '3日均價(元)', '情感欄位']]
y_df3 = df3['當日漲跌']

# 分割數據集為訓練集和測試集
X_train_df2, X_test_df2, y_train_df2, y_test_df2 = train_test_split(X_df2, y_df2, test_size=0.1, random_state=42)
X_train_df3, X_test_df3, y_train_df3, y_test_df3 = train_test_split(X_df3, y_df3, test_size=0.1, random_state=42)

# 創建隨機森林模型
rf_clf_df2 = RandomForestClassifier(random_state=42)
rf_clf_df3 = RandomForestClassifier(random_state=42)

# 訓練模型
rf_clf_df2.fit(X_train_df2, y_train_df2)
rf_clf_df3.fit(X_train_df3, y_train_df3)

# 預測測試集
y_pred_df2 = rf_clf_df2.predict(X_test_df2)
y_pred_df3 = rf_clf_df3.predict(X_test_df3)

# 評估模型
accuracy_df2 = accuracy_score(y_test_df2, y_pred_df2)
accuracy_df3 = accuracy_score(y_test_df3, y_pred_df3)

print("df2 隨機森林模型準確度:", accuracy_df2)
print("df3 隨機森林模型準確度:", accuracy_df3)

# 顯示混淆矩陣和分類報告（可選）
#print("df2 混淆矩陣:\n", confusion_matrix(y_test_df2, y_pred_df2))
#print("df2 分類報告:\n", classification_report(y_test_df2, y_pred_df2))

#print("df3 混淆矩陣:\n", confusion_matrix(y_test_df3, y_pred_df3))
#print("df3 分類報告:\n", classification_report(y_test_df3, y_pred_df3))
```
